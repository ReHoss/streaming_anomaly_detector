```python
df_joined = (df_displays
 .set_index('impressionId')
 .join(other=df_clicks.set_index('impressionId'),
                                 on='impressionId',
                                 rsuffix='_clicks'))
```

```python
(df_joined
 .groupby([df_joined.timestamp.dt.minute, 'uid'])
 .count()
 
 .assign(rate=lambda df: df.uid_clicks/df.eventType)
 .sort_values(by='rate',
              ascending=False)
 .head(25)
)
```
